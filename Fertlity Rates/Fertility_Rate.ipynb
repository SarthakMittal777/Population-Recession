```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os



# Extract the years (column names)
years = df.columns[1:]

# Create a directory to store the outlier plots
if not os.path.exists("outlier_plots"):
    os.makedirs("outlier_plots")

# Define a function to identify outliers based on IQR
def find_outliers_iqr(data, threshold=1.5):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    return (data < lower_bound) | (data > upper_bound)

# Loop through each row (country) and create outlier plots
for i in range(len(df)):
    row_data = df.iloc[i]
    country_name = row_data[0]
    fertility_data = row_data[1:].astype(float)

    # Find outliers in fertility data using IQR
    is_outlier = find_outliers_iqr(fertility_data)

    # Create a line plot with outliers highlighted
    plt.figure(figsize=(12, 6))
    plt.plot(years, fertility_data, marker='o', linestyle='-', label='Fertility Data', color='blue')
    plt.scatter(years[is_outlier], fertility_data[is_outlier], marker='x', c='red', s=100, label='Outliers')
    plt.title(f'Fertility Rate for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend()
    plt.grid(True)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Save the plot as an image in the "outlier_plots" folder
    plt.tight_layout()
    plt.savefig(f'outlier_plots/outlier_plot_{country_name}.png')
    plt.close()  # Close the current plot to free up memory
```
