```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os



# Extract the years (column names)
years = df.columns[1:]

# Define a list of statistics you want to calculate
statistics = ["Mean", "Variance", "Mode", "Skewness", "Kurtosis"]

# Create a directory to store the images
output_folder = "statistics_plots"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through each statistical measure
for stat in statistics:
    # Create a subfolder for the statistic
    stat_folder = os.path.join(output_folder, stat)
    if not os.path.exists(stat_folder):
        os.makedirs(stat_folder)

    # Create an empty list to store statistic values
    stat_values = []

    # Loop through each row (country) and calculate the statistic
    for i in range(len(df)):
        row_data = df.iloc[i]
        fertility_data = row_data[1:].astype(float)

        if stat == "Mean":
            stat_value = fertility_data.mean()
        elif stat == "Variance":
            stat_value = fertility_data.var()
        elif stat == "Mode":
            stat_value = fertility_data.mode().values[0]
        elif stat == "Skewness":
            stat_value = fertility_data.skew()
        elif stat == "Kurtosis":
            stat_value = fertility_data.kurtosis()

        # Append the statistic value to the list
        stat_values.append(stat_value)

    # Create a line plot to show all statistic values
    plt.figure(figsize=(12, 6))
    plt.plot(stat_values, marker='o', linestyle='-')
    plt.title(f'{stat} for All Countries')
    plt.xlabel('Country Index')
    plt.ylabel(stat)
    plt.grid(True)

    # Save the combined plot as an image in the subfolder
    plt.tight_layout()
    plt.savefig(f'{stat_folder}/combined_{stat.lower()}_plot.png')
    plt.close()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress



# Extract the years (column names)
years = df.columns[1:]

# Loop through each row and create a separate plot with linear regression curve
for i in range(len(df)):
    row_data = df.iloc[i]
    country_name = row_data[0]
    fertility_data = row_data[1:].astype(float)
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(range(len(years)), fertility_data)
    regression_line = slope * np.array(range(len(years))) + intercept
    
    # Create a line plot with linear regression curve
    plt.figure(figsize=(12, 6))
    plt.plot(years, fertility_data, marker='o', linestyle='-', label='Fertility Data')
    plt.plot(years, regression_line, linestyle='--', label='Linear Regression', color='red')
    plt.title(f'Total Fertility Rate for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend()
    plt.grid(True)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Save the plot as an image (you can specify a directory)
    plt.tight_layout()
    plt.savefig(f'fertility_regression_plot_{country_name}.png')
    plt.close()  # Close the current plot to free up memory
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.cluster import KMeans
import os

# Load the CSV file into a Pandas DataFrame


# Extract the years (column names)
years = df.columns[1:]

# Create a directory to store the K-Means clustering plots
if not os.path.exists("kmeans_clustering_plots"):
    os.makedirs("kmeans_clustering_plots")

# Loop through each row and create a separate plot with K-Means clustering
for i in range(len(df)):
    row_data = df.iloc[i]
    country_name = row_data[0]
    fertility_data = row_data[1:].astype(float)
    
    # Create a K-Means clustering model with a specified number of clusters (e.g., 3)
    num_clusters = 3
    kmeans = KMeans(n_clusters=num_clusters)
    
    # Fit the model to the fertility data
    fertility_data = fertility_data.values.reshape(-1, 1)  # Reshape fertility_data
    kmeans.fit(fertility_data)
    
    # Get cluster labels for each year
    cluster_labels = kmeans.labels_
    
    # Create a line plot with different colors for each cluster
    plt.figure(figsize=(12, 6))
    for cluster_id in range(num_clusters):
        cluster_data = fertility_data[cluster_labels == cluster_id]
        cluster_years = years[:len(cluster_data)]  # Adjust years to match the data
        plt.plot(cluster_years, cluster_data, marker='o', linestyle='-', label=f'Cluster {cluster_id + 1}')
    plt.title(f'Total Fertility Rate Clustering for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend()
    plt.grid(True)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Save the plot as an image in the "kmeans_clustering_plots" folder
    plt.tight_layout()
    plt.savefig(f'kmeans_clustering_plots/kmeans_clustering_plot_{country_name}.png')
    plt.close()  # Close the current plot to free up memory
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import os

# Load the CSV file into a Pandas DataFrame


# Extract the years (column names)
years = df.columns[1:]

# Create a directory to store the images and anomaly detection results
if not os.path.exists("anomaly_detection_plots"):
    os.makedirs("anomaly_detection_plots")

# Loop through each row and perform anomaly detection with Isolation Forest
for i in range(len(df)):
    row_data = df.iloc[i]
    country_name = row_data[0]
    fertility_data = row_data[1:].astype(float)

    # Create an Isolation Forest model
    isolation_forest = IsolationForest(contamination=0.05)  # Adjust the contamination parameter as needed

    # Fit the model to the data and predict anomalies (-1 for anomalies, 1 for inliers)
    anomalies = isolation_forest.fit_predict(fertility_data.values.reshape(-1, 1))

    # Create a line plot with anomalies highlighted
    plt.figure(figsize=(12, 6))
    plt.plot(years, fertility_data, marker='o', linestyle='-', label='Fertility Data')
    plt.scatter(years[anomalies == -1], fertility_data[anomalies == -1], c='red', marker='x', label='Anomalies')
    plt.title(f'Anomaly Detection for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend()
    plt.grid(True)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Save the plot as an image in the "anomaly_detection_plots" folder
    plt.tight_layout()
    plt.savefig(f'anomaly_detection_plots/anomaly_detection_plot_{country_name}.png')
    plt.close()  # Close the current plot to free up memory
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the CSV file into a Pandas DataFrame

# Extract the years (column names)
years = df.columns[1:]

# Create a directory to store the images and SVD results
if not os.path.exists("svd_plots"):
    os.makedirs("svd_plots")

# Loop through each row and perform SVD
for i in range(len(df)):
    row_data = df.iloc[i]
    country_name = row_data[0]
    fertility_data = row_data[1:].astype(float)

    # Perform SVD on the fertility data
    U, S, Vt = np.linalg.svd(fertility_data.values.reshape(-1, 1), full_matrices=False)

    # Reconstruct the data using a subset of singular values (e.g., first 3)
    num_singular_values = 3
    reconstructed_data = U[:, :num_singular_values] @ np.diag(S[:num_singular_values]) @ Vt[:num_singular_values, :]

    # Create a line plot with the reconstructed data
    plt.figure(figsize=(12, 6))
    plt.plot(years, fertility_data, marker='o', linestyle='-', label='Fertility Data')
    plt.plot(years, reconstructed_data, linestyle='--', label=f'SVD Reconstruction (K={num_singular_values})', color='red')
    plt.title(f'SVD for {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Fertility Rate')
    plt.legend()
    plt.grid(True)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Save the plot as an image in the "svd_plots" folder
    plt.tight_layout()
    plt.savefig(f'svd_plots/svd_plot_{country_name}.png')
    plt.close()  # Close the current plot to free up memory
```
